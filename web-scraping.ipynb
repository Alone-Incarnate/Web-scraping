```python
import requests
from bs4 import BeautifulSoup
```

```python

# URL of the article
url = "https://techcrunch.com/2024/04/30/sams-clubs-ai-powered-exit-tech-reaches-20-of-stores/"

# Send a GET request to fetch the webpage content
response = requests.get(url)

# Check if the request was successful
if response.status_code == 200:
    print("Successfully fetched the webpage!")

    # Parse the page content using BeautifulSoup
    soup = BeautifulSoup(response.content, 'html.parser')

    # Extract the article's title (Handle potential changes in class or structure)
    title_tag = soup.find('h1')  # Removed class since it might be dynamic
    title = title_tag.get_text() if title_tag else "Title not found"

    # Extract the article's author (Handle changes in class or tag)
    author_tag = soup.find('div', class_='article__byline')
    author = author_tag.get_text().strip() if author_tag else "Author not found"

    # Extract the article's publication date (Handle changes in structure)
    date_tag = soup.find('time')
    date = date_tag.get_text() if date_tag else "Date not found"

    # Extract the article's content paragraphs (Handle potential changes)
    article_content = soup.find_all('p')
    article_text = '\n'.join([p.get_text() for p in article_content if p.get_text()])

    # Print the extracted content
    print(f"Title: {title}")
    print(f"Author: {author}")
    print(f"Published Date: {date}\n")
    print("Article Content:\n")
    print(article_text)
else:
    print(f"Failed to fetch the webpage. Status code: {response.status_code}")
```
